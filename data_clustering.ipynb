```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.read_csv('weather.csv')
```

```python
wind_precipitation_data = data[['Data.Wind.Speed', 'Data.Precipitation']]
temperature_precipitation_data = data[['Data.Temperature.Avg Temp', 'Data.Precipitation']]
```

```python
#Function to perform KMeans clustering and plot results
def perform_clustering_and_plot(data, title):
    #Perform KMeans clustering with 3 clusters
    kmeans = KMeans(n_clusters=3, n_init=10)
    kmeans.fit(data)
    labels = kmeans.labels_

    #Plotting the clusters
    #'X's represent the centroids of the clusters.
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100)
    plt.title(title)
    plt.xlabel(data.columns[0])
    plt.ylabel(data.columns[1])
    plt.show()

#Cluster for Wind and Precipitation
perform_clustering_and_plot(wind_precipitation_data, 'Clustering: Wind vs Precipitation')
#Cluster for Temperature and Precipitation
perform_clustering_and_plot(temperature_precipitation_data, 'Clustering: Temperature vs Precipitation')
```
